```julia
using DataFrames, PyPlot;
c₁="#838300";
c₂="#008383";
c₃="#830083";
```

```julia
data = readtable("data.dat",separator='\t',header=true);
data[:Max_Energy] = log10(data[:Max_Energy]);
data[:EL_Gate] = log10(data[:EL_Gate]);
display(data)
```

```julia


fig = figure("Static Error",figsize=(12,8)) # Create a new figure
ax = axes();
p2 = plot(data[:N],data[:Max_Energy],color=c₁,linewidth=2,linestyle="-",marker="s");
p1 = plot(data[:N],data[:EL_Gate],color=c₂,linewidth=2,linestyle="--",marker="o");
#ax[:set_xlim]([0,20]);
#ax[:set_ylim]([-5,0]);
ylabel(L"Timing error statistic [$\log_{10}(\epsilon(N))$]");
xlabel("Oversampling factor [N]");
title("Timing Static Error Plot");
legend(["Max Energy","EL Gate"]);

fig[:canvas][:draw]()
```
